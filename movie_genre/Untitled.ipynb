```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\heman\OneDrive\Desktop\AFTER BTECH\innomatics\ml\Movie_kind\Movie_Raw_Data.csv")
```

```python
df.drop(columns=['Unnamed: 0','id'],inplace=True)
```

```python
df
```

```python
df['len']=df['description'].str.len()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
X=df['description']
```

```python
y=df['genre']
```

```python
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
X=X.apply(lambda x:re.sub('[^A-Za-z\s]',"",x))
```

```python
X=X.str.lower()
```

```python
X
```

```python
Stopwords=stopwords.words('english')
X=X.apply(lambda x:" ".join([i for i in nltk.word_tokenize(x) if i not in Stopwords]))
```

```python
X
```

```python
lemmatizer=WordNetLemmatizer()
X=X.apply(lambda x:" ".join([lemmatizer.lemmatize(i) for i in nltk.word_tokenize(x)]))
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=99)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer()
X_train_trans = vect.fit_transform(X_train)
X_test_trans = vect.transform(X_test)
```

```python
X_train_trans
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier
from sklearn. tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
base_estimator = DecisionTreeClassifier(max_depth=1)
# AdaBoost classifier
ada_clf = AdaBoostClassifier(estimator=base_estimator,
n_estimators=50,learning_rate=1.0,random_state=42)
ada_clf.fit(X_train_trans,y_train)
y_pred=ada_clf.predict(X_test_trans)
print('Accuracy:',accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
